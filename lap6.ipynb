```python
# Question 1

numbers = []

for i in range(5):
    num = float(input("Enter a number: "))
    numbers.append(num)

total = sum(numbers)
average = total / len(numbers)
maximum = max(numbers)
minimum = min(numbers)

print("Sum:", total)
print("Average:", average)
print("Maximum:", maximum)
print("Minimum:", minimum)
```

```python
# Question 2

words = []

for i in range(5):
    word = input("Enter a word: ")
    words.append(word)

words_tuple = tuple(words)

print("First word:", words_tuple[0])
print("Last word:", words_tuple[-1])
print("Total words:", len(words_tuple))

search_word = input("Enter a word to search: ")

if search_word in words_tuple:
    print("Word is in the tuple")
else:
    print("Word is not in the tuple")
```

```python
# Question 3

students = {}

for i in range(3):
    name = input("Enter student name: ")
    score = float(input("Enter student score: "))
    students[name] = score

print("Students and scores:")
for name, score in students.items():
    print(name, ":", score)

highest_student = max(students, key=students.get)

print("Student with highest score:", highest_student)
print("Highest score:", students[highest_student])
```

```python
# Question 4

list1 = input("Enter first list of numbers separated by spaces: ")
list2 = input("Enter second list of numbers separated by spaces: ")

set1 = set(map(int, list1.split()))
set2 = set(map(int, list2.split()))

print("Union:", set1.union(set2))
print("Intersection:", set1.intersection(set2))
print("Difference:", set1.difference(set2))
```

```python
# Question 5

names = []

for i in range(5):
    name = input("Enter a name: ")
    names.append(name)

names_tuple = tuple(names)
print("Tuple:", names_tuple)

name_dict = {}

for name in names:
    name_dict[name] = len(name)

print("Dictionary:", name_dict)

length_set = set(name_dict.values())
print("Set of unique name lengths:", length_set)
```
