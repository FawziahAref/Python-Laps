```python
# Question 1
import math

def circle_area(radius):
    return math.pi * radius * radius

circumference = lambda r: 2 * math.pi * r

radius = float(input("Enter radius of the circle: "))

area = circle_area(radius)
circ = circumference(radius)

print("Area =", round(area, 2))
print("Circumference =", round(circ, 2))
```

```python
# Question 2
import re

def sum_numbers(nums):
    return sum(nums)

text = input("Enter a string: ")

numbers = re.findall(r'\d+', text)
numbers = list(map(int, numbers))

total = sum_numbers(numbers)
double = lambda x: x * 2

print("Numbers found:", numbers)
print("Sum:", total)
print("Doubled value:", double(total))
```

```python
# Question 3
import math

nums_input = input("Enter numbers separated by commas: ")
nums_list = list(map(float, nums_input.split(",")))

def square_list(lst):
    return list(map(lambda x: x**2, lst))

squares = square_list(nums_list)

max_square = max(squares)
sqrt_sum = math.sqrt(sum(squares))

print("Squares:", squares)
print("Maximum square:", max_square)
print("Square root of sum of squares:", sqrt_sum)
```

```python
# Question 4
import re
import math

password = input("Enter password: ")

upper = re.search(r'[A-Z]', password)
lower = re.search(r'[a-z]', password)
number = re.search(r'\d', password)

if upper and lower and number:
    print("Password is strong")
else:
    print("Password is weak")

scores_input = input("Enter 5 scores separated by spaces: ")
scores = list(map(float, scores_input.split()))

def average(lst):
    return sum(lst) / len(lst)

avg = average(scores)
rounded_avg = math.ceil(avg)

avg_lambda = lambda x: x  # lambda used as requested

print("Rounded average score:", rounded_avg)
```

```python
# Question 5
import re
import math

paragraph = input("Enter a paragraph: ")

vowel_words = re.findall(r'\b[AEIOUaeiou]\w*', paragraph)

def count_vowels(text):
    return len(re.findall(r'[AEIOUaeiou]', text))

total_vowels = count_vowels(paragraph)

lengths = list(map(lambda w: len(w), vowel_words))

if lengths:
    avg_length = round(sum(lengths) / len(lengths), 2)
else:
    avg_length = 0

print("Words starting with vowels:", vowel_words)
print("Total vowels:", total_vowels)
print("Average vowel-word length:", avg_length)
```
