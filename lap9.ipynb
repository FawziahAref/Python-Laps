```python
grades = np.array([[87, 96, 70],
                   [100, 87, 90],
                   [94, 77, 90],
                   [100, 81, 82]])

print("sum =", np.sum(grades))
print("min =", np.min(grades))
print("max =", np.max(grades))
print("mean =", np.mean(grades))
print("std =", np.std(grades))
print("var =", np.var(grades))

print("mean per column =", np.mean(grades, axis=0))
print("mean per row =", np.mean(grades, axis=1))
print("max per row =", np.max(grades, axis=1))
```

```python
import numpy as np

A = np.array([[10, 20, 30],
              [40, 50, 60]])

B = np.array([1, 2, 3])

# Add B to A
add_result = A + B
print("A + B =\n", add_result)

# Multiply A by 2
mul_result = A * 2
print("A * 2 =\n", mul_result)

# Subtract 5 from A
sub_result = A - 5
print("A - 5 =\n", sub_result)
```

```python
grades = np.array([[70, 85, 90],
                   [88, 76, 95],
                   [60, 80, 72]])

# Index of largest value in whole array
max_index = np.argmax(grades)
print("Index of largest value (flattened):", max_index)

# Index of highest grade in each row
row_max_index = np.argmax(grades, axis=1)
print("Index of max in each row:", row_max_index)

# Index of lowest grade in each column
col_min_index = np.argmin(grades, axis=0)
print("Index of min in each column:", col_min_index)
```

```python
import numpy as np

# إنشاء المصفوفة
table = np.array([
    [5, 8, 2, 7],
    [1, 9, 4, 3],
    [6, 0, 11, 10],
    [12, 13, 14, 15]
])

print("Table =\n", table)

# الصف الثاني (R2)
print("Second row =", table[1])

# العمود الثالث (C3)
print("Third column =", table[:, 2])

# sub-table (R2–R3 و C2–C3)
print("Sub-table =\n", table[1:3, 1:3])
```

```python
scores = np.array([
    [70, 80, 90],
    [60, 75, 85],
    [88, 92, 79],
    [55, 65, 72]
])

print("Original scores =\n", scores)

# زيادة درجات الفيزياء +5 (العمود الثاني)
scores[:, 1] = scores[:, 1] + 5

# استبدال درجات أول طالب
scores[0] = [100, 100, 100]

# جعل عمود CS (الأخير) = 0
scores[:, 2] = 0

print("Updated scores =\n", scores)
```

```python
import numpy as np

x = np.array([2, 0, 1, 5, 4, 1, 9])
A = np.array([[3, 8, 1],
              [6, 2, 9]])

# Part 1: argsort
indices = np.argsort(x)
sorted_x = x[indices]

print(indices)
print(sorted_x)

# Part 2: min location
min_index = np.unravel_index(A.argmin(), A.shape)
print(min_index)

# Part 3: where
indices_where = np.where(x > 3)
print(indices_where)

# Part 4: split
parts = np.split(x, [2, 5])
for part in parts:
    print(part)

# Part 5: expand and squeeze
expanded = np.expand_dims(x, axis=0)
print(expanded)

squeezed = np.squeeze(expanded)
print(squeezed)
```

```python
# Argsort
x = np.array([4, 2, 7, 1, 5])

sorted_indices = np.argsort(x)
print(sorted_indices)
print(x[sorted_indices])

# شرح:
# np.argsort(x) لا يرجع القيم مرتبة مباشرة،
# بل يرجع فهارس العناصر بعد الترتيب.
# باستخدام هذه الفهارس نستطيع عرض المصفوفة بشكل مرتب.


# Minimum Location
A = np.array([[8, 3, 6],
              [1, 9, 4]])

min_index = A.argmin()
location = np.unravel_index(min_index, A.shape)

print(min_index)
print(location)

# شرح:
# A.argmin() يبحث عن أصغر قيمة في المصفوفة
# ويرجع موقعها كرقم واحد.
# np.unravel_index يحول هذا الرقم إلى صف وعمود.


# Where
x = np.array([1, 4, 2, 6, 3, 7])

result = np.where(x > 3)
print(result)

# شرح:
# np.where(x > 3) يبحث عن العناصر
# التي تحقق الشرط (أكبر من 3)
# ثم يعرض أماكنها داخل المصفوفة.


# Split
x = np.array([10, 20, 30, 40, 50, 60, 70])

parts = np.split(x, [2, 5])

print(parts)

# شرح:
# np.split يقسم المصفوفة إلى أجزاء
# حسب المواقع المحددة.
# هنا تم تقسيمها إلى:
# أول عنصرين، ثم 3 عناصر، ثم الباقي.


# Expand & Squeeze
x = np.array([1, 2, 3, 4, 5, 6, 7])

expanded = np.expand_dims(x, axis=0)
print(expanded.shape)

squeezed = np.squeeze(expanded)
print(squeezed.shape)

# شرح:
# np.expand_dims يضيف بُعد جديد للمصفوفة
# مثل تحويلها من صف عادي إلى صف داخل مصفوفة.
# np.squeeze يحذف الأبعاد الزائدة
# ويرجعها إلى شكلها الطبيعي.
```
